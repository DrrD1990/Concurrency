```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mathtext as mathtext

mathtext.FontConstantsBase.sup1 = 0.25
plt.rcParams["font.family"] = "sans-serif"
plt.rcParams["font.sans-serif"] = "Arial"
plt.rcParams['mathtext.fontset'] = "custom"
plt.rcParams["mathtext.rm"] = "Arial"
plt.rcParams['mathtext.it'] = 'Arial:italic'

np.random.seed(16)
  
fig, ax1 = plt.subplots(figsize=(10, 7))

a = 0.3
b = 0.5
N = 10
M = 10
N0 = np.random.randint(0, 2) 
M0 = np.random.randint(2, 4) 

samplepaths = 1

s = np.zeros((samplepaths, N))
X = np.zeros((samplepaths, N))
t = np.zeros((samplepaths, M))
Y = np.zeros((samplepaths, M))

X[:,0] = N0
Y[:,0] = M0

for j in range(samplepaths):
    for i in range(N-1):
        if N0 == 0:
            h = - np.log(np.random.rand())/a
            s[j,i+1] = s[j,i] + h
            X[j,i+1] = N0
            N0 = 1
        elif N0 == 1:
            h = - np.log(np.random.rand())/b
            s[j,i+1] = s[j,i] + h
            X[j,i+1] = N0
            N0 = 0

for j in range(samplepaths):
    for i in range(M-1):
        if M0 == 2:
            h = - np.log(np.random.rand())/a
            t[j,i+1] = t[j,i] + h
            Y[j,i+1] = M0
            M0 = 3
        elif M0 == 3:
            h = - np.log(np.random.rand())/b
            t[j,i+1] = t[j,i] + h
            Y[j,i+1] = M0
            M0 = 2
        
## Generates plots
for r in range(samplepaths):
    ax1.step(s[r,:], X[r,:], where='pre', color='b')
    ax1.step(t[r,:], Y[r,:], where='pre', color='b')

    
#ax.set_xlabel('Time', fontsize=14)
#ax.set_ylabel('Events observed', fontsize=14)
y1 = [-2,-2,-2,-2,-1,-2,-1,-2,-2,-2,-2,-2,-1,-2]
y2 = [-3,-3,-4,-3,-3,-3,-3,-3,-4,-3,-4,-3,-3,-3]
x = [0.,          1.29572309,  1.91088705,  3.28426455,  7.01317639,
  8.09598772,  9.45987415, 11.78059267, 12.85863665, 15.85907787, 16.39782523,
 17.1021163,  18.84966735, 20.05476249]
ax1.step(x, y1, color='b')
ax1.step(x, y2, color='b')
ax1.set_xlim(0, 20)
ax1.set_ylim(-4.5, 3.5)
#plt.xlim(0, 20)
ax1.set_xlabel('time', fontsize=28)
ax1.set_xticks([])
ax1.set_yticks([-4, -3, -2, -1, 0, 1, 2, 3])
ax1.set_yticklabels(['inactive', 'active', 'inactive', 'active', '$\it{l}$', '$\it{h}$', '$\it{l}$', '$\it{h}$'], fontsize=20)
#plt.xlabel('time', fontsize=20)
#plt.xticks([])
#plt.yticks([-4, -3, -2, -1, 0, 1, 2, 3], ['inactive', 'active', 'inactive', 'active', 'l', 'h', 'l', 'h'], fontsize=20)
#plt.tight_layout()
#plt.legend(loc=2)
#ax2 = ax1.twinx()
#ax2.set_ylim(-4.5, 3.5)
#ax2.set_yticks([-3.8, -3.2, -1.8, -1.2, 0.5, 2.5])
#ax2.set_yticklabels(['in model 3', 'edge (1, 2)', 'in model 2', 'edge (1, 2)', 'node 2', 'node 1'], fontsize=20)plt.savefig('model_2&3.png', bbox_inches = 'tight')
#fig.text(0.1, 0.833, 'h', ha='center', va='center', fontsize=20, family='Arial', style='italic')
#fig.text(0.1, 0.735, 'l', ha='center', va='center', fontsize=20, family='Arial', style='italic')
#fig.text(0.1, 0.645, 'h', ha='center', va='center', fontsize=20, family='Arial', style='italic')
#fig.text(0.1, 0.547, 'l', ha='center', va='center', fontsize=20, family='Arial', style='italic')
fig.text(-0.1, 0.78, 'node 1', ha='center', va='center', fontsize=20)
fig.text(-0.1, 0.59, 'node 2', ha='center', va='center', fontsize=20)
fig.text(-0.1, 0.4, 'edge (1, 2)\nin model 2', ha='center', va='center', fontsize=20)
fig.text(-0.1, 0.21, 'edge (1, 2)\nin model 3', ha='center', va='center', fontsize=20)
plt.savefig('models2&3.pdf', bbox_inches="tight")
plt.show()
#print(s[r,:], X[r,:])
#print(t[r,:], Y[r,:])
a = np.array(s[r,:])
a = np.append(a, t[r,:])
a.sort()
#print(a)
```
